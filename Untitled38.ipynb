```python
probabilitas = 0.05  # Probabilitas untuk setiap pixel menjadi noise
noise = np.random.rand(image.shape[0], image.shape[1])
image_with_noise = image.copy()
image_with_noise[noise < probabilitas / 2] = 0
image_with_noise[noise > 1 - probabilitas / 2] = 255
```

```python
#random crop
height, width = image.shape[:2]
```

```python
crop_size = 250
```

```python
x = np.random.randint(0, width - crop_size)
y = np.random.randint(0, height - crop_size)
```

```python
cropped_image_random = image[y:y+crop_size, x:x+crop_size]
```

```python
cv2.imshow('Original Image', image)
cv2.imshow('Random Crop', cropped_image_random)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#center crop
height, width = image.shape[:2]
```

```python
crop_size = 250
```

```python
start_x = (width - crop_size) // 2
start_y = (height - crop_size) // 2
```

```python
cropped_image_center = image[start_y:start_y + crop_size, start_x:start_x + crop_size]
```

```python
cv2.imshow('Original Image', image)
cv2.imshow('Center Crop', cropped_image_center)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
